## Generating a corpus 

A corpus is generated according to the following process:

* $N \sim Pois(\psi)$
* $\theta_d \sim Dir(\alpha) \propto \prod_{i=1}^{T}\theta_{d,i}^{\alpha_i - 1}$ with $T$ the number of topics
* $z \vert \theta_d \sim Mult(\theta_d) \propto \prod_{i=1}^{T} \theta_{d,i}^{z_{i}}$
* $w \vert \theta_d, z \sim Mult(\beta) \propto \prod_{i=1}^{V} \beta_{i}^{w_{i}}$ with $V$ the vocabulary size.

Full likelihood:

\begin{split}
p(\mathbf{\theta}, \mathbf{w}, \mathbf{z} \vert \alpha, \beta) & = \prod_{d=1}^M \prod_{n=1}^N p(\theta_d, w_n, z_n \vert \alpha, \beta) \\
 & = \prod_{d=1}^M p(\theta_d \vert \alpha) \prod_{n=1}^N p(w_{d,n} \vert z_{d,n}, \beta)p(z_{d,n} \vert \theta_d) \\
\end{split}

```python
psi = 10  # Avg. number of words per document
V = 10  # Vocabulary size
T = 5  # Number of beta

M = 100  # Number of documents
N = np.random.poisson(psi, M)  # Number of words per document

alpha = [1./T]*T  # Dirichlet prior on beta
gamma = [1./V]*V  # Dirichlet prior on words
theta = np.random.dirichlet(alpha, M)  # Distribution of beta per document
beta = np.random.dirichlet(gamma, T)  # Distribution of words per beta

docs = []
for docid in range(M):
  doc = []
  for word in range(N[docid]):
    z = np.random.multinomial(1, theta[docid])
    w = np.nonzero(np.random.multinomial(1, beta[np.nonzero(z)[0][0]]))[0][0]
    doc.append(w)
  docs.append(doc)
```

## Full conditionals 



\begin{split}
p(\theta_d \vert \mathbf{w}_d, \mathbf{z}_d, \alpha, \beta) & =  \frac{p(\theta, \mathbf{w}_d, \mathbf{z}_d \vert \alpha, \beta)}{p(\mathbf{w}_d, \mathbf{z}_d \vert \alpha, \beta)} \\
 & \propto p(\mathbf{\theta}_d \vert \alpha)p(\mathbf{z}_d \vert \theta_d) \\
 & \propto \prod_{i=1}^{T}\theta_{d,i}^{\alpha_i - 1}  \prod_{n=1}^{N}\prod_{i=1}^{T} \theta_{d,i}^{z_{d,n,i}} \\
 & = \prod_{i=1}^{T} \theta_{d,i}^{\alpha_i - 1 + \bar{z_{d,i}}}
\end{split}
Hence, $\theta_d \sim Dir(\alpha_i + \bar{z_{d,i}})$ with $\bar{z_{d,i}}$ the number of words belonging to topic $i$ in document $d$.

\begin{split}
p(z_{d, n} \vert \theta_d, w_{d, n}, \alpha, \beta) & =  \frac{p(\theta_d, w_{d, n}, z_{d, n} \vert \alpha, \beta)}{p(\theta_d, w_{d,n} \vert \alpha, \beta)} \\
 & \propto p(w_{d,n} \vert z_{d,n}, \beta)p(z_{d,n} \vert \theta_d) \\
 & \propto \prod_{j=1}^{V} \beta_{d, n, j}^{w_{d,n,j}} \prod_{i=1}^{T} \theta_{d,i}^{z_{d, n, i}} \\
\end{split}

Hence $z_{d,n} \sim Mult(\beta_{w_{d,n}}*\theta_d)$ where $*$ denotes the element-wise multiplication.



## Gibbs sampling

```python
%%time

def pz_s(w, theta_s):
    p = beta[:, w]*theta[d]
    return p / np.sum(p)
    
theta_s = np.random.dirichlet(alpha, M)
nb_iter = 1000
theta_est = np.zeros((nb_iter, M, T))
for it in range(nb_iter):
    for d in range(M):
        z_s = np.zeros((N[d], T))
        for i in range(N[d]):
            z_s[i, :] = np.random.multinomial(1, pz_s(docs[d][i], theta_s))
        z_bar = np.sum(z_s, 0)
        theta_s[d] = np.random.dirichlet(alpha + z_bar, 1)
    theta_est[it, :, :] = theta_s
```

```python
docid = 30
print "Estimated = %s" % ','.join("%.3f" % x for x in np.mean(theta_est[:, docid, :], axis=0))
print "Real      = %s" % ','.join("%.3f" % x for x in theta[docid])
```
