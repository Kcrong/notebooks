# Bayesian estimation of a logistic regression

```python
import numpy as np
import scipy.stats

%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # To plot the likelihood in 3D
from matplotlib import cm
```

```python
logistic = lambda x: 1. / (1. + np.exp(-x))
```

## 1. Fake Dataset

### Generate

Sample a fake dataset from the following generative process:
* $x \sim \mathcal{U}(-1, 1)$
* $y \sim \mathcal{Ber}(\sigma(x))$ with $\sigma(x) = \frac{1}{1+e^{-x}}$

```python
N = 1000  # A dataset of 1000 points
x = np.array([np.r_[1., np.random.uniform(-1, 1, 2)] for i in range(N)])  # (N, 3) matrix
w = np.array([[2, -6, 5]]) #  (1, 3) Arbitrary directionfor the logistic.
y = np.array([np.random.rand() < p for p in logistic(x.dot(w.T))])  # (N, 1) Sample from a bernoulli
```

### Visualize

```python
def apply_to_grid(f, bias=1., xlim=(-1, 1), ylim=(-1, 1), s=100):
  """ Apply a function of 3 parameters to a 2D grid discretization.
  
  The functions to be evaluated (either the logistic or the likelihood
  in our case) depend on three parameters but only two parameters are of 
  interest. The third parameter is taken to be constant.
  
  Args:
      f: the function to apply to the grid.
      bias: a constant for the first parameter.
      xlim: range for the second parameter.
      ylim: range for the third parameter.
      s: number of points per dimension.
  
  Returns:
      X: matrix of x coordinates
      Y: matrix of y coordinates
      Z: matrix of evauations of f at the gridpoints.
  """
  X, Y = np.array(np.meshgrid(np.linspace(xlim[0], xlim[1], s),
                              np.linspace(ylim[0], ylim[1], s)))
  z = np.hstack((bias*np.ones((s*s, 1)),
                 np.reshape(X, (s*s, 1)),
                 np.reshape(Y, (s*s, 1))))
  return X, Y, np.reshape(f(z), (s,s))
```

```python
def sample_plot(x, y, bg=None):
  """ Visualize the dataset. 
  
  If the function is given a $bg$ parameter it plots a
  nice background with separator and probability density
  for the data.
  
  Args:
      x: (np.array) (N, 3) array of coordiantes.
      y: (np.array) (N, 1) array of labels.
      bg: tuple retrun by the apply_to_grid function
  """
  fig = plt.figure(figsize=(6, 6))
  fig.suptitle(r"Visualization of the fake dataset", fontsize=18)
  ax = fig.add_subplot(1, 1, 1)
  ax.set_aspect(1)
  ax.set_xlabel(r'$x_1$', fontsize=18)
  ax.set_ylabel(r'$x_2$', fontsize=18)
  ax.set_axisbelow(True)
  ax.set_xlim(-1, 1)
  ax.set_ylim(-1, 1)
  ax.autoscale(False)

  if bg is not None:
    X, Y, Z = bg
    ax.pcolor(X, Y, Z, cmap=cm.RdBu, vmin=0, vmax=1)
    ax.contour(X, Y, Z, levels=[0.5], vmin=-1, vmax=1)
  ax.scatter(x[:, 1], x[:, 2], c=y, vmin=-1, vmax=1)
  plt.show()
```

```python
X, Y, Z = apply_to_grid(lambda x: logistic(x.dot(w.T)))
sample_plot(x, y, bg=(X, Y, Z))
```

## 2. Likelihood

```python
def lik(beta):
  """ Returns the evaluation of the likelihood for a given beta.
  
  Args:
      beta: (np.array) (N, K) array of parameters to evaluate the 
            likelihood at.
  
  Returns:
      (np.array) (N, 1) array of evaluations of the likelihood for
      every given parameters.
  """
  p = logistic(x.dot(beta.T))
  return np.array([np.prod(y*p + (1-y)*(1-p), 0)]).T
```

```python
def likelihood_plot(X, Y, Z):
  
  # 3D version
  fig = plt.figure(figsize=(6, 12))
  fig.suptitle('Likelihood', fontsize=18)
  ax = fig.add_subplot(2, 1, 1, projection='3d')
  ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.RdBu,linewidth=0,
                        antialiased=False)
  
  # Contour plot
  ax = fig.add_subplot(2, 1, 2)
  ax.set_xlabel(r'$\beta_1$', fontsize=18)
  ax.set_ylabel(r'$\beta_2$', fontsize=18)
  ax.pcolor(X, Y, Z, cmap=cm.RdBu)
  ax.contour(X, Y, Z)
  ax.grid(True)
  plt.show()
```

```python
X, Y, Z = apply_to_grid(lik, bias=2., xlim=(-7, -5), ylim=(4, 6))
likelihood_plot(X, Y, Z)
```

## 3. Metropolis Hastings

```python
rnorm = lambda mu, sigma, N: np.random.multivariate_normal(mu, sigma, N)

def MetropolisHastingsSampler(N, target=lik, scale=0.1):
  """ Sample N times using MH with a symmetric gaussian proposal.
  
  Args:
      N: The number of samples to draw
      target: the target function to sample from
      scale: scale parameter for the variance of the proposal
  Returns:
      samples: np.array (N, 3) array of samples
      ratios: np.array (N, 1) array of $\frac{f(x^{t+1})}{f(x^t)}
      rate: average acceptance rate.
  """
  f = target
  samples = np.zeros((N, 3))
  ratios = np.zeros((N, 1))
  rate = 0.

  x_prev = np.random.rand(1,3)
  for i in xrange(N):
    x_cur = x_prev + rnorm([0]*3, scale*np.eye(3, 3), 1)
    ratio = f(x_cur)/f(x_prev)
    if np.random.uniform(0, 1) < ratio:
      x_prev = x_cur
      rate += 1
    samples[i] = x_prev[0]
    ratios[i] = ratio[0]
  rate /= N

  return samples, ratios, rate
```

```python
samples, ratios, rate = MetropolisHastingsSampler(5000, target=lik)
```

```python
print "Acceptance Rate =", rate
```

```python
estimators = np.cumsum(np.array(samples), axis=0)/np.array([range(1, len(samples)+1)]).T
figsize(12, 6)
plt.plot(estimators)
plt.title("Convergence of estimators")
plt.xlabel("Number of iterations")
plt.ylabel("estimation")
plt.show()
```

## Gradient descent

```python
def grad(theta):
  return np.sum((logistic(x.dot(theta.T)) - y) * x, 0)
```

```python
N = 100
theta = np.random.rand(1, 3)
estimators = np.zeros((N, 3))
for t in range(N):
  theta -= 1./(50 + t) * grad(theta)
  estimators[t] = theta
```

```python
print "Estimated =", theta[0]
print "Real =", w[0]
```

```python
plt.plot(estimators)
plt.title("Convergence of the gradient descent")
plt.xlabel("Number of iterations")
plt.ylabel("estimated parameters")
plt.show()
```
