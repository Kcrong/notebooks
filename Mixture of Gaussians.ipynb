```python
import numpy as np
import scipy.stats

import matplotlib.pyplot as plt

bern = lambda p: scipy.stats.bernoulli(p)
norm = lambda mu, sigma: scipy.stats.multivariate_normal(mean=mu, cov=sigma)
```

## Generate a fake dataset

* $z \sim \mathcal{Ber}(p)$
* $x \mid z \sim \mathcal{N}(\mu_z, \sigma^2)$
* $\mu_1 \sim \mathcal{N}(0, \sigma^2)$
* $\mu_2 \sim \mathcal{N}(0, \sigma^2)$


```python
N = 500
p = 0.8
mu_1 = [-1, 1]
mu_2 = [1, -1]
sigma = .5*np.eye(2, 2)
```

```python
z = np.array([bern(p).rvs(N)]).T
x = z*norm(mu_1, sigma).rvs(N) + (1-z)*norm(mu_2, sigma).rvs(N)
```

```python
fig = plt.figure(figsize=(6, 6))
fig.suptitle(r"Visualization of the fake dataset", fontsize=18)
ax = fig.add_subplot(1, 1, 1)
ax.set_aspect(1)
ax.set_xlabel(r'$x_1$', fontsize=18)
ax.set_ylabel(r'$x_2$', fontsize=18)
ax.set_axisbelow(True)
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.scatter(x[:, 0], x[:, 1], c=z)
plt.show()
```

## Complete data posterior

\begin{split}
\pi (z, \mu_1, \mu_2 \mid x) & \propto \pi (x, z, \mu_1, \mu_2) \\
&= \pi (x\mid z) \pi (z)\pi (\mu_1)\pi (\mu_2) \\
&\propto \exp(-\frac{\mu_1^2}{2\sigma^2})\exp(-\frac{\mu_2^2}{2\sigma^2})\prod_{k} \left(p \exp(-\frac{(x_k - \mu_1)^2}{2\sigma^2})\right)^{z_k} \left((1-p)\exp(-\frac{(x_k -\mu_2)^2}{2\sigma^2})\right)^{1-z_k}
\end{split}

## Get the full conditionals

Full conditionals

\begin{split}
\pi (z \mid x, \mu_1, \mu_2) & = \frac{\pi (z, \mu_1, \mu_2 \mid x)}{\pi (\mu_1, \mu_2 \mid x)} \\
& = \frac{\pi (z, \mu_1, \mu_2 \mid x)}{\sum_z \pi (z, \mu_1, \mu_2 \mid x)}
\end{split}


\begin{split}
\sum_z \pi (z, \mu_1, \mu_2 \mid x) &\propto \sum_{z}\left( \exp(-\frac{\mu_1^2}{2\sigma^2})\exp(-\frac{\mu_2^2}{2\sigma^2})\prod_{k} \left(p \exp(-\frac{(x_k - \mu_1)^2}{2\sigma^2})\right)^{z_k} \left((1-p)\exp(-\frac{(x_k -\mu_2)^2}{2\sigma^2})\right)^{1-z_k} \right) \\
& = \exp(-\frac{\mu_1^2}{2\sigma^2})\exp(-\frac{\mu_2^2}{2\sigma^2}) \prod_{k} \sum_{z_k} \left( \left(p \exp(-\frac{(x_k - \mu_1)^2}{2\sigma^2})\right)^{z_k} \left((1-p)\exp(-\frac{(x_k -\mu_2)^2}{2\sigma^2})\right)^{1-z_k} \right) \\ 
& = \exp(-\frac{\mu_1^2}{2\sigma^2})\exp(-\frac{\mu_2^2}{2\sigma^2}) \prod_{k} \left( p \exp(-\frac{(x_k - \mu_1)^2}{2\sigma^2}) + (1-p)\exp(-\frac{(x_k -\mu_2)^2}{2\sigma^2}) \right)
\end{split}

and

\begin{split}
\pi (Z = z \mid x, \mu_1, \mu_2) & = \prod_{k} \frac{\left(p \exp(-\frac{(x_k - \mu_1)^2}{2\sigma^2})\right)^{z_k} \left((1-p)\exp(-\frac{(x_k -\mu_2)^2}{2\sigma^2})\right)^{1-z_k}}{ \left( p \exp(-\frac{(x_k - \mu_1)^2}{2\sigma^2}) + (1-p)\exp(-\frac{(x_k -\mu_2)^2}{2\sigma^2}) \right)} \\
& = \prod_{k} {P}(Z_k=z_k \mid x_k, \mu_1, \mu_2)
\end{split}

\begin{split}
\pi (\mu_1 \mid x, z, \mu_2) & = \frac{\pi (z, \mu_1, \mu_2 \mid x)}{\pi (z, \mu_2 \mid x)}
\end{split}


Rather than explicitly computing $\pi (z, \mu_2 \mid x)$ it suffices to select the appropriate terms in $\pi (z, \mu_1, \mu_2 \mid x)$ and infer the normalization constant by identifying a known density.

\begin{split}
\pi (\mu_1\mid x, z, \mu_2) &\propto \exp(-\frac{\mu_1^2}{2\sigma^2})\prod_{k} \left(\exp(-\frac{(x_k - \mu_1)^2}{2\sigma^2})\right)^{z_k} \\
&= \exp\left(-\frac{1}{{2\sigma^2}}\left(\sum_{z_k=1} (x_k^2 - 2x_k\mu_1 + \mu_1^2) +\mu_1^2\right)\right) \\
&= \exp\left(-\frac{(1+\sum_{z_k=1} 1)}{{2\sigma^2}}\left(\mu_1^2 -\frac{2\mu_1\sum_{z_k=1}x_k}{(1+\sum_{z_k=1} 1)} - \frac{\sum_{z_k=1}x_k^2}{(1+\sum_{z_k=1} 1)}) \right)\right) \\
&\propto \exp\left(-\frac{(1+\sum_{z_k=1} 1)}{{2\sigma^2}}\left(\mu_1 -\frac{\sum_{z_k=1}x_k}{(1+\sum_{z_k=1} 1)} \right)^2\right) 
\end{split}

Hence,
\begin{split}
\pi(\mu_1 \mid x, z, \mu_2) \sim \mathcal{N}(\frac{\sum_{z_k=1}x_k}{(1+\sum_{z_k=1} 1)}, \frac{\sigma^2}{(1+\sum_{z_k=1} 1)})
\end{split}



## Gibbs sampling

```python
def pz_given_x(mu_1, mu_2):
  pos = p * np.exp(np.sum((x - mu_1)*(x - mu_1), 1) / sigma[0,0]**2)
  neg = (1-p) * np.exp(np.sum((x - mu_2)*(x - mu_2), 1) / sigma[0,0]**2)
  return neg/(pos+neg)

def GibbsSampler(N):
  mu_1_post = []
  mu_2_post = []
  mu_1s = -1 + 2*np.random.rand(1, 2) # random initial point
  mu_2s = -1 + 2*np.random.rand(1, 2)
  for i in range(N):
    pz = pz_given_x(mu_1s, mu_2s)
    zs = np.array([[bern(p).rvs(1)[0] for p in pz]]).T
    mu_1s = np.sum(zs*x, 0) / (1 + np.sum(zs))
    mu_2s = np.sum((1-zs)*x, 0) / (1 + np.sum(1-zs))
    sigma_1s = sigma / (1 + np.sum(zs))
    sigma_2s = sigma / (1 + np.sum(1-zs))
    mu_1_post.append(norm(mu_1s, sigma_1s).rvs(1))
    mu_2_post.append(norm(mu_2s, sigma_2s).rvs(1))

  return np.array(mu_1_post), np.array(mu_2_post)
```

```python
%time mu_1_post, mu_2_post = GibbsSampler(100) # mega lent
```

```python
print "mu_1est = %s (real = %s)" % (np.mean(mu_1_post, 0), mu_1)
print "mu_2est = %s (real = %s)" % (np.mean(mu_2_post, 0), mu_2)
```

```python
figsize(12, 6)
f, (ax1, ax2) = plt.subplots(2, sharex=True, sharey=True)
f.suptitle(r"Convergence of samples for $\mu_1$ and $\mu_2$", fontsize=18)
ax1.plot(mu_1_post)
ax1.set_ylabel("$\mu_1$", fontsize=14)
ax2.plot(mu_2_post)
ax2.set_ylabel("$\mu_2$", fontsize=14)
plt.show()
```

```python
figsize(12, 6)
f, (ax1, ax2) = plt.subplots(2, sharex=True, sharey=True)
f.suptitle(r"Hisotgrams for $\mu_1$ and $\mu_2$", fontsize=18)
ax1.hist(mu_1_post, bins=50)
ax1.set_xlabel("$\mu_1$", fontsize=14)
ax2.hist(mu_2_post, bins=50)
ax2.set_xlabel("$\mu_2$", fontsize=14)
plt.show()
```
